```python
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import numpy as np
from sklearn.linear_model import LinearRegression


```

```python
df=pd.read_csv('/content/sample_data/Boston (1).csv')
```

```python
df
```

```python
x=df.drop("medv",axis='columns')
```

```python
x
```

```python
y=df.medv
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
lm=LinearRegression()
```

```python
model=lm.fit(x_train,y_train)
```

```python
ytrainpred=lm.predict(x_train)
```

```python
ytestpred=lm.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
MSE = np.square(np.subtract(y_test,ytestpred)).mean()
```
