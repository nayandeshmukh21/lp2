```python
q1 = df['sepal_length'].quantile(0.25)
q2 = df['sepal_length'].quantile(0.50)
q3 = df['sepal_length'].quantile(0.75)
```

```python
print(q1,q2,q3)
```

```python
IQR = q3 - q1
IQR
```

```python
UL = q3 + (1.5 * IQR)
LL = q1 - (1.5 * IQR)
print(UL,LL)
```

```python
outl_z = pd.DataFrame(np.where((df['sepal_length']>UL)))
outl_z
```

```python
outl1_z = pd.DataFrame(np.where((df['sepal_length']<LL)))
outl1_z
```

```python
to_replace = pd.DataFrame(np.where(to_replace[0]<LL,df['sepal_length'].mean(),to_replace[0]))
to_replace
```

```python
#capping and flooring
q_10th = df['sepal_length'].quantile(0.1)
q_90th = df['sepal_length'].quantile(0.9)

temp = df.copy()

print(q_10th,q_90th)

for i in np.where(temp['sepal_length']>UL)[0]:
    if(temp['sepal_length'][i]<q_10th):
        temp.loc[:,('sepal_length')] = q_10th
    else:
        temp.loc[:,('sepal_length')] = q_90th
temp
```
