```python
q1 = df['Total'].quantile(0.25)
q2 = df['Total'].quantile(0.50)
q3 = df['Total'].quantile(0.75)
```

```python
IQR = q3 - q1
IQR
```

```python
UL = q3 + (1.5 * IQR)
LL = q1 - (1.5 * IQR)
print(UL,LL)
```

```python
def normalize(values):
    return (values - values.min())/(values.max() - values.min()) 
```

```python
df1 = df.copy()

cols = ['FE_mark','SE_mark','TE_mark','BE_mark']

df1[cols] = df1[cols].apply(normalize)
df1
```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()

cols = ['FE_mark','SE_mark','TE_mark','BE_mark','Total']

df2 = df.copy()

df2[cols] = mms.fit_transform(df[cols])

df2
```

```python
# flooring and capping
q_10th = df['Total'].quantile(0.1)
q_90th = df['Total'].quantile(0.9)

temp = df.copy()

print(q_10th,q_90th)

for i in np.where(temp['Total']>UL)[0]:
    if(temp['Total'][i]<q_10th):
        temp.loc[:,('Total')] = q_10th
    else:
        temp.loc[:,('Total')] = q_90th
temp
```
